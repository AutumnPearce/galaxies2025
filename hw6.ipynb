```python
from astropy.cosmology import Planck18 as cosmo
sf_gxy = pd.read_csv(f'hw2_data/DESI_SF.dat', sep =' ', names = ["lambd", "f_lambd", "f_error", "model"])
sf_gxy["f_lambd"] = sf_gxy["f_lambd"] * 1e-17
z_sf_gxy = .031
sf_gxy["lambd_rest"] = sf_gxy["lambd"] / (1+z_sf_gxy) # de-redshift the lambdas
sf_gxy["f_lambd_rest"] = sf_gxy["f_lambd"] * (1+z_sf_gxy) # de-redshift the f_lambda 
D_L = cosmo.luminosity_distance(z_sf_gxy).to('cm').value  # [cm]
sf_gxy["L_lambd_rest"] = sf_gxy["f_lambd_rest"] * 4 * np.pi * D_L**2 
sf_error = sf_gxy['f_error'] * (1+z_sf_gxy) * 4 * np.pi * D_L**2

ages = np.logspace(-2, 1, 20)
M_stars = np.logspace(8,10,20)

age_grid, mass_grid = np.meshgrid(ages, M_stars, indexing='ij')
chi2_grid = np.zeros(age_grid.shape)

sp = fsps.StellarPopulation(imf_type=0, sfh=0)
for i in range(age_grid.shape[0]):
    for j in range(age_grid.shape[1]):
        age = age_grid[i, j]
        mass = mass_grid[i, j]

        wave, spec = sp.get_spectrum(tage=age, peraa=True)
        model_L = spec * L_solar * mass

        model_interp = np.interp(sf_gxy["lambd_rest"], wave, model_L)
        chi2 = np.sum(((sf_gxy["L_lambd_rest"] - model_interp) / sf_error) ** 2)

        chi2_grid[i, j] = chi2

print(chi2_grid)

from matplotlib.colors import LogNorm
plt.figure(figsize=(8,6))
im = plt.imshow(chi2_grid, origin='lower', 
                extent=[M_stars[0], M_stars[-1], ages[0], ages[-1]],
                aspect='auto', cmap='viridis', norm=LogNorm())
plt.colorbar(im, label=r'$\chi^2$')
plt.xlabel('Stellar Mass [$M_\odot$]')
plt.ylabel('Age [Gyr]')
plt.xscale('log')
plt.yscale('log')
plt.title('Chi-squared over Age and Mass grid')
plt.show()
```
